```python
import pandas as pd
theme_list = pd.read_csv('theme_new.csv')['theme'].tolist()
```

```python
from langchain.embeddings import OpenAIEmbeddings
embedding = OpenAIEmbeddings(
    model='text-embedding-ada-002',
    deployment = 'text-embedding-ada-002',
    openai_api_key ="xxx",
    chunk_size=1,
)
documents = theme_list
embeddings = embedding.embed_documents(documents)
```

```python
import pickle
with open('data.pkl', 'wb') as file:
    pickle.dump(embeddings, file)
# 读的时候用这个
# with open('data.pkl', 'rb') as file:
#     loaded_embeddings = pickle.load(file)
```

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=50, random_state=0)
kmeans.fit(embeddings)
labels = kmeans.labels_
```

```python
from sklearn.cluster import AffinityPropagation
affinity_propagation = AffinityPropagation(random_state=42)
labels = affinity_propagation.fit_predict(embeddings)
```

```python
from sklearn.metrics import pairwise_distances_argmin_min
centroids = affinity_propagation.cluster_centers_
closest_embeddings_indices, _ = pairwise_distances_argmin_min(centroids, embeddings)
representative_words = [documents[index] for index in closest_embeddings_indices]
```

```python
clusters = {}
for i, label in enumerate(labels):
    if label not in clusters:
        clusters[label] = []
    clusters[label].append(theme_list[i])
df = pd.DataFrame({
    "Representative Word": representative_words,
    "Group Members": [clusters[i] for i in range(len(representative_words))]
})
df
```
